```python
# numpy choice 

import numpy as np
import matplotlib.pyplot as plt

# using choice method 
gfg = np.random.choice(13,5000)

count,bins,ignored = plt.hist(gfg,25,density=False)

plt.show()

arr1= np.random.choice(12,5000)

count,bins,ignored  = plt.hist(arr1,20,density=True)

plt.show()
```
